```python
# Bài 3

import numpy as np

arr = np.array([[88, 35, 62], [11, 98, 34],[64, 24, 48]])

min_val = np.min(arr)
max_val = np.max(arr)
sum_val = np.sum(arr)
mean_val = np.mean(arr)
std_val = np.std(arr)

sum_col = np.sum(arr, axis=0)
mean_col = np.mean(arr, axis=0)

sum_row = np.sum(arr, axis=1)
mean_row = np.mean(arr, axis=1)

print("Min:", min_val)
print("Max:", max_val)
print("Sum:", sum_val)
print("Mean:", mean_val)
print("STD:", std_val)

print("\nSumOfColumn:", sum_col)
print("MeanOfColumn:", mean_col)

print("\nSumOfRow:", sum_row)
print("MeanOfRow:", mean_row)
```

```python
# Bài 5

import numpy as np
import time

arr = np.random.randint(1, 21, size=10)
print("Mảng ban đầu:", arr)

square = arr ** 2
sqrt_arr = np.sqrt(arr)

print("\nBình phương (vectorized):", square)
print("Căn bậc hai (vectorized):", sqrt_arr)

add_result = np.add(arr, 5)        
mul_result = np.multiply(arr, 2)   
sqrt_result = np.sqrt(arr)         

print("\nKết quả dùng np.add(arr, 5):", add_result)
print("Kết quả dùng np.multiply(arr, 2):", mul_result)
print("Kết quả dùng np.sqrt(arr):", sqrt_result)

arr2d = np.random.randint(1, 21, size=(1000, 1000))

start_vec = time.time()
vec_result = arr2d * 2
end_vec = time.time()

start_loop = time.time()
loop_result = np.zeros_like(arr2d)

for i in range(arr2d.shape[0]):
    for j in range(arr2d.shape[1]):
        loop_result[i, j] = arr2d[i, j] * 2

end_loop = time.time()

print("\n⏱ Thời gian vectorization :", end_vec - start_vec)
print("⏱ Thời gian vòng lặp thủ công:", end_loop - start_loop)
```

```python
# Bài 6

import numpy as np

arr = np.array([[3, 7, 4, 5], [3, 5, 2, 4],[6, 3, 9, 7], [4, 8, 3, 5]])

sum_col = np.sum(arr, axis=0)
mean_col = np.mean(arr, axis=0)
std_col = np.std(arr, axis=0)

sum_row = np.sum(arr, axis=1)
mean_row = np.mean(arr, axis=1)
std_row = np.std(arr, axis=1)

print("Tổng theo cột (axis=0):", sum_col)
print("Trung bình theo cột (axis=0):", mean_col)
print("Độ lệch chuẩn theo cột (axis=0):", std_col)

print("\nTổng theo hàng (axis=1):", sum_row)
print("Trung bình theo hàng (axis=1):", mean_row)
print("Độ lệch chuẩn theo hàng (axis=1):", std_row)
```

```python
# Bài 7

import numpy as np
import time

A = np.array([
    [8, 12, 14, 6, 4],
    [18, 13, 2, 2, 12],
    [6, 18, 5, 10, 10],
    [9, 2, 4, 12, 12],
    [2, 3, 2, 17, 3]
])

b = np.array([10, 20, 30, 40, 50])

add_result = A + b
sub_result = A - b

print("Kết quả cộng (broadcasting):\n", add_result, "\n")
print("Kết quả trừ (broadcasting):\n", sub_result, "\n")

sqrt_result = np.sqrt(A)
add5_result = np.add(A, 5)

print("Kết quả căn bậc hai (ufunc):\n", sqrt_result, "\n")
print("Kết quả cộng 5 vào mảng (ufunc):\n", add5_result, "\n")

start = time.time()
manual = np.zeros_like(A)

for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        manual[i, j] = A[i, j] + 5

end = time.time()
print("Thời gian thực thi vòng lặp thủ công:", round(end - start, 6), "giây")
```

```python
# Bài 8

import numpy as np

A = np.array([
    [6, 9, 7],
    [2, 1, 5],
    [7, 8, 3]
])
A_inv = np.linalg.inv(A)
print("Ma trận nghịch đảo:\n", A_inv, "\n")
eig_vals, eig_vecs = np.linalg.eig(A)

print("Giá trị riêng của ma trận:", eig_vals, "\n")
print("Véc tơ riêng của ma trận:\n", eig_vecs)
```
