```python
# Bài 1

lst = [1, 2, 3, 2, 5, 1, 7, 9, 9, 4]

even_positions = lst[0::2]

tpl = tuple(lst)

d = {i: lst[i] for i in range(len(lst))}

s = set(lst)
removed = len(lst) - len(s)

print("Các vị trí chẵn:", even_positions)
print("Tuple:", tpl)
print("Dict:", d)
print("Set:", s)
print("Số phần tử bị loại trùng:", removed)
```

```python
# Bài 4

import pandas as pd

score_series = pd.Series([7.5, 8.0, 6.5, 9.0, 8.5])
print(score_series)

name = ['An','Bình','Châu','Dũng','Hà']
age = [20,21,22,20,23]
score = [7.5, 8.0, 6.5, 9.0, 8.5]

df = pd.DataFrame({
    "name": name,
    "age": age,
    "score": score
})

df.info()
print(df.describe())

filtered = df.loc[df["score"] >= 8, :]
print("Sinh viên >= 8 điểm:")
print(filtered.values.tolist())
```

```python
# Bài 5

import pandas as pd

df_students = pd.read_csv("students.csv", names=["id", "name"])
df_scores = pd.read_csv("scores.csv", names=["id", "score"])

df = pd.merge(df_students, df_scores, on="id")

report = df.groupby(["id", "name"])["score"].agg(
    score_min="min",
    score_max="max",
    score_mean="mean"
).reset_index()

print(report)

report.to_csv("report.csv", index=False)
```

```python
# Bài 6

import numpy as np

a = np.array([[1], [2], [3], [4]])      
b = np.array([[10, 20, 30, 40, 50]])   

c = a * b
print("Kết quả a*b:\n", c)

sqrt_c = np.sqrt(c)
log_c = np.log(c)
print("sqrt =\n", sqrt_c)
print("log =\n", log_c)

mean_axis1 = c.mean(axis=1)
min_axis1 = c.min(axis=1)
max_axis1 = c.max(axis=1)
std_axis1 = c.std(axis=1)

print("mean theo axis=1 =", mean_axis1)
print("min theo axis=1 =", min_axis1)
print("max theo axis=1 =", max_axis1)
print("std theo axis=1 =", std_axis1)

```
