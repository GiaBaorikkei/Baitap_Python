```python
# Bài 2

import csv

with open('students.csv', mode='r', encoding='utf-8') as infile:
    reader = csv.reader(infile)
    header = next(reader)
    
    high_grade_students = []

    for row in reader:
        name, age, grade = row[0], row[1], float(row[2])
        if grade >= 7:
            high_grade_students.append(row)
            print(f"Sinh viên: {name}, Tuổi: {age}, Điểm: {grade}")

with open('high_grade_students.csv', mode='w', newline='', encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(header) 
    writer.writerows(high_grade_students) 
print("Đã lưu sinh viên có điểm >= 7 vào file")
```

```python
# Bài 3

import json
with open('user_data.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)
    for user in data:
        if user['id'] == 3:
            print(f"Tên: {user['name']}, Tuổi: {user['age']}")
new_user = {
    "id": 4,
    "name": "Nguyen Van D",
    "age": 30
}
data.append(new_user)
with open('user_data.json', 'w', encoding='utf-8') as json_file:
    json.dump(data, json_file, ensure_ascii=False, indent=4)
print("Đã thêm người dùng mới vào file JSON")
```

```python
# Bài 6
import json

with open('inventory.json', 'r', encoding='utf-8') as json_file:
    inventory = json.load(json_file)
new_product = {
    "product_id": 3,
    "name": "Tablet",
    "quantity": 20,
    "price": 150000
}
inventory.append(new_product)
with open('inventory.json', 'w', encoding='utf-8') as json_file:
    json.dump(inventory, json_file, ensure_ascii=False, indent=4)
print("Đã thêm sản phẩm mới vào file inventory.json")
```

```python
# Bài 8

from datetime import datetime, date

today = date.today()

with open('log.txt', 'r', encoding='utf-8') as file:
    for line in file:
        timestamp_str = line.split(' - ')[0]
        try:
            timestamp = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
            if timestamp.date() == today:
                print(line.strip())
        except ValueError:
            continue
```
