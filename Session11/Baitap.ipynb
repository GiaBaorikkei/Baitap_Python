```python
# Bài 1

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 1)

y = x**2 + 2*x

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set_title("Biểu đồ đầu tay của Gia Bảo")

ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")

ax.grid(True, linestyle="--")

plt.show()
```

```python
# Bài 3

import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

fig, ax = plt.subplots(figsize=(8, 6))

ax.hist(data, bins=30)

ax.set_title("Biểu đồ Histogram với dữ liệu phân phối chuẩn")
ax.set_xlabel("Giá trị")
ax.set_ylabel("Tần suất")

plt.show()
```

```python
# Bài 4

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2
y3 = x**3

fig, ax = plt.subplots()

ax.plot(x, y1, color='red', label='y = x')
ax.plot(x, y2, color='green', label='y = x^2')
ax.plot(x, y3, color='blue', label='y = x^3')

ax.set_title("Biểu đồ nhiều đường")
ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")

ax.legend()

plt.show()
```

```python
# Bài 5

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2

scatter_x = np.random.randint(0, 50, 20)
scatter_y = np.random.randint(0, 50, 20)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title("y = x")

axs[0, 1].plot(x, y2, color="red")
axs[0, 1].set_title("y = x²")

axs[1, 0].scatter(scatter_x, scatter_y, color="green")
axs[1, 0].set_title("Scatterplot 1")

axs[1, 1].scatter(np.random.randn(20), np.random.randn(20), color="purple")
axs[1, 1].set_title("Scatterplot 2")

plt.tight_layout()

plt.show()
```

```python
# Bài 6

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y, label="y = sin(x)")

x_point = np.pi / 2
y_point = np.sin(x_point)

ax.annotate(
    "Đỉnh sóng\n(x = π/2)",
    xy=(x_point, y_point),         
    xytext=(x_point + 0.5, y_point + 0.5), 
    arrowprops=dict(arrowstyle="->") 
)

ax.set_title("Đồ thị y = sin(x) với chú thích")
ax.set_xlabel("Trục x")
ax.set_ylabel("Trục y")

plt.show()
```

```python
# Bài 7

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.rand(10, 10)

fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(
    data,
    cmap="viridis",  
    annot=True,       
    fmt=".2f",        
    ax=ax
)

ax.set_title("Heatmap từ ma trận dữ liệu ngẫu nhiên")
ax.set_xlabel("Cột")
ax.set_ylabel("Hàng")

plt.show()
```

```python
# Bài 8

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x = np.linspace(0, 2 * np.pi, 100)
fig, ax = plt.subplots(figsize=(10, 6))
line, = ax.plot(x, np.sin(x), color='blue')

ax.set_ylim(-1.5, 1.5)
ax.set_title('Đồ thị động y = sin(x + t)')
ax.set_xlabel('Trục x')
ax.set_ylabel('Trục y')
ax.grid(True, linestyle='--', alpha=0.7)

def update(frame):
    y = np.sin(x + 0.1 * frame)
    line.set_ydata(y)
    return line,

ani = FuncAnimation(fig, update, frames=200, interval=30, blit=True, repeat=True)

plt.show()
```
